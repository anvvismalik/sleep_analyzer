```python
#hyperparameter tuning
estimators=[10,50,100,150,200,300]
depths=[5,10,15,20,None]

train_scores = {depth: [] for depth in depths}
test_scores = {depth: [] for depth in depths}

#calculating for different estimators and depths
for n in estimators:
  for depth in depths:
    model=RandomForestRegressor(n_estimators=n,max_depth=depth,random_state=30)
    model.fit(X_train,y_train)
    trained=model.score(X_train,y_train)
    tested=model.score(X_test,y_test)
    train_scores[depth].append(trained)
    test_scores[depth].append(tested)
```

```python
#plotting
plt.figure(figsize=(10,6))
for depth in depths:
  #for each depth there will be a different graph
  #for each depth the y axis will be the test scores for that depth
  #the x axis will be the corresponding estimator values
  plt.plot(estimators,test_scores[depth],label=f'max-depth is{depth}',marker='o',alpha=0.7)


plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('R² Score on Test Set')
plt.title('Effect of n_estimators and max_depth on Model Performance')
plt.legend()
plt.show()
```

```python
#GRAPH ANALYSIS
#from the graph we can interpret that a very good r2 test score can be achieved by 100 trees
#and even higher in 150 trees,after that the increase is not significant
#we have to choose between 100 and 150 for estimators
#for 100 also the score is considerably high and more trees means more time for computation
#so lets go for 100 trees
#for choosing depth,need to make a choice between none and 10 for optimal result
```

```python
#choosing between depth=10 and depth=none
#comparing the test and train scores for both
index_100 = 2
test_10=test_scores[10][index_100]
train_10=train_scores[10][index_100]

test_none=test_scores[None][index_100]
train_none=train_scores[None][index_100]

print(f"Depth = 10 -> Train Score: {train_10:.5f}, Test Score: {test_10:.5f}")
print(f"Depth = None -> Train Score: {train_none:.5f}, Test Score: {test_none:.5f}")
```

```python
from sklearn.model_selection import cross_val_score
#cross validation
model_depth_10 = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
model_depth_none = RandomForestRegressor(n_estimators=100, max_depth=None, random_state=42)

cv_scores_10 = cross_val_score(model_depth_10, X_train, y_train, cv=5, scoring='r2')
cv_scores_none = cross_val_score(model_depth_none, X_train, y_train, cv=5, scoring='r2')

print(f"Mean CV Score (max_depth=10): {np.mean(cv_scores_10):.5f}")
print(f"Mean CV Score (max_depth=None): {np.mean(cv_scores_none):.5f}")
```

```python
#the results for both depth=10 and none are very high and almost similar
#since depth=none is only very very slightly higher than depth=10
#this means that depth=none is not creating a considerable difference
#and will only contibute to more computational time
#also cross validation score for depth=10 is more
#so lets choose depth=10
```

```python
#FINAL CHOSEN FOR RANDOM FOREST REGRESSOR
#max-depth is 10 and n_estimators is 100
final_model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=30)
final_model.fit(X_train, y_train)

train_score = final_model.score(X_train, y_train)
test_score = final_model.score(X_test, y_test)

print(f"Final Model -> Train Score: {train_score:.5f}, Test Score: {test_score:.5f}")
```

```python
#creating thresholds using a dynamic approach of
#using percentiles as well as guidelines for health

def basic_threshold(df):
  thresholds={
      'Heart_Rate_Variability':np.percentile(df['Heart_Rate_Variability'],25),#25 was chosen because here low hrv leads to poor sleep quality
      'Body_TemperatureMin':36.1,
      'Body_TemperatureMax':37.2,
      'Movement_During_Sleep':np.percentile(df['Movement_During_Sleep'],85),
      'Sleep_Duration_HoursMin':7,
      'Sleep_Duration_HoursMax':9,
      'Caffeine_Intake_mg':min(300,np.percentile(df['Caffeine_Intake_mg'],85)),
      'Stress_Level':np.percentile(df['Stress_Level'],85),
      'Bedtime_Consistency':np.percentile(df['Bedtime_Consistency'],25),
      'Light_Exposure_hours':np.percentile(df['Light_Exposure_hours'],85)

  }
  return thresholds

  min_thresholds = {
    'Heart_Rate_Variability': 20,
    'Body_TemperatureMin': 35.5,
    'Body_TemperatureMax': 36.5,
    'Movement_During_Sleep': 0,
    'Sleep_Duration_HoursMin': 4,
    'Sleep_Duration_HoursMax': 5,
    'Caffeine_Intake_mg': 0,
    'Stress_Level': 0,
    'Bedtime_Consistency': 0,
    'Light_Exposure_hours': 0
}

max_thresholds = {
    'Heart_Rate_Variability': 150,
    'Body_TemperatureMin': 36.9,
    'Body_TemperatureMax': 38.0,
    'Movement_During_Sleep': 10,
    'Sleep_Duration_HoursMin': 6,
    'Sleep_Duration_HoursMax': 10,
    'Caffeine_Intake_mg': 400,
    'Stress_Level': 100,
    'Bedtime_Consistency': 60,
    'Light_Exposure_hours': 10
}
```

```python
def update_thresholds(X_train, thresholds, shap_values):
    new_thresh = thresholds.copy()
    feature_names = list(X_train.columns)

    for i in range(len(feature_names)):
        feat = feature_names[i]

        if feat in new_thresh:
            mean_shap = np.mean(shap_values[:, i])

            if mean_shap < -0.1:
                new_thresh[feat] *= 0.9
            elif mean_shap > 0.1:
                new_thresh[feat] *= 1.1

            # Proper clipping to min/max bounds
            new_thresh[feat] = np.clip(
                new_thresh[feat],
                min_thresholds.get(feat, -np.inf),
                max_thresholds.get(feat, np.inf)
            )

    return new_thresh

```

```python
def generate_recommendations(user_data, thresholds):
    recs = []

    # Normalize user keys to lowercase for consistent comparison
    user_data_normalized = {k.lower(): v for k, v in user_data.items()}
    thresholds_normalized = {k.lower(): v for k, v in thresholds.items()}

    # Handle body temperature and sleep separately as they have min and max
    bt = user_data_normalized.get('body_temperature')
    if bt is not None:
        bt_min = thresholds_normalized.get('body_temperaturemin')
        bt_max = thresholds_normalized.get('body_temperaturemax')
        if bt_min is not None and bt < bt_min:
            recs.append(f"Increase Body Temperature from {bt} to ≥{round(bt_min, 2)}")
        if bt_max is not None and bt > bt_max:
            recs.append(f"Reduce Body Temperature from {bt} to ≤{round(bt_max, 2)}")

    sleep = user_data_normalized.get('sleep_hours')
    if sleep is not None:
        sleep_min = thresholds_normalized.get('sleep_duration_hoursmin')
        sleep_max = thresholds_normalized.get('sleep_duration_hoursmax')
        if sleep_min is not None and sleep < sleep_min:
            recs.append(f"Increase Sleep Hours from {sleep} to ≥{round(sleep_min, 2)}")
        if sleep_max is not None and sleep > sleep_max:
            recs.append(f"Reduce Sleep Hours from {sleep} to ≤{round(sleep_max, 2)}")

    # Now check the rest of the features
    for feature, value in user_data_normalized.items():
        if feature in ['body_temperature', 'sleep_hours']:
            continue  # already handled above
        threshold = thresholds_normalized.get(feature)
        if threshold is None:
            continue
        if value < threshold:
            recs.append(f"Increase {feature.replace('_', ' ').title()} from {value} to ≥{round(threshold, 2)}")
        elif value > threshold:
            recs.append(f"Reduce {feature.replace('_', ' ').title()} from {value} to ≤{round(threshold, 2)}")

    if not recs:
        recs.append(" All values are within healthy/personalized ranges!")

    return recs

```
